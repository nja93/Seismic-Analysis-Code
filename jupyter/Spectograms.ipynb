<B>This is a notebook to showcase the creation of Spectograms</B>for earthquake and non-earthquake signals
<p>What is a  spectogram?</p> A spectrogram is a visual representation of the frequency content of a seismogram as it changes with time. 

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from obspy import UTCDateTime
from obspy.clients.fdsn import Client
client = Client("IRIS")
```

We will consider comparing singals from and earthquake and an explosion

```python
endt = UTCDateTime("2014-07-11T06:17:50")
startt = UTCDateTime("2014-07-11T06:15:50")
location = "00"
client = Client("IRIS")
network = "IU"
channel = "BHZ"
station = "TUC"
```

```python
st = client.get_waveforms(network, station, location, channel, startt, endt)
```

After running the code to request the seismogram, I wi;; extract a seismic trace from the stream and then plot it

```python
tr = st[0]
```

```python
fig = tr.plot()
plt.rcParams['figure.figsize'] = (12, 4) #change size to be app same as seimogram
fig = tr.spectrogram(per_lap=.5, wlen=2)
```
